# Sorting experiments
## Run the experiment
**Parameters:**
- n: number of items to sort
- niters: number of independent trials for each experiment
- predGenID: prediction setting $\in$ {"class", "decay"}
- algosToTest: list of algorithms to test from:
    - BH: Sort using Binary Heap
    - FH: Sort using Fibonacci Heap
    - SL: Sort using Skip List with offline rank predictions
    - SLR: Sort using Skip List with online rank predictions
    - DC: Sort using Skip List with Dirty Comparisons
